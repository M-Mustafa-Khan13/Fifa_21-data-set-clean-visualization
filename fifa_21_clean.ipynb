```python
age_perf = data.groupby("Age")["Overall Rating"].mean()

plt.plot(age_perf.index, age_perf.values)
plt.xlabel("Age")
plt.ylabel("Average Overall Rating")
plt.title("Average Player Performance by Age")
plt.show()
```

```python
top_countries = (
    data["Nationality"].value_counts()
      .head(20)
)

plt.barh(top_countries.index, top_countries.values)
plt.xticks(rotation=70)
plt.ylabel("Number of Players")
plt.title("Top 15 Player-Producing Countries")
plt.show()
```

```python


fig,ax=plt.subplots(1,2,figsize=(10,6))


data = data[data["Height(Cms)"] >=0]

height_value=data["Height(Cms)"]
weight_value=data["Weight(Kgs)"]


ax[0].hist(height_value,bins=20,color="red",edgecolor="black",label="Height Info")
ax[0].set_title("Height Distribution of Players")
ax[0].set_xlabel("Height Range of Players (CMs)")
ax[0].set_ylabel("Frequency")
ax[0].legend()


ax[1].hist(weight_value,bins=20,color="green",edgecolor="black",label="Weight Info")
ax[1].set_title("Weight Distribution of Players")
ax[1].set_xlabel("Weight Range of Players (KGs)")
ax[1].set_ylabel("Frequency")
ax[1].legend()


fig.suptitle("Player Distribution by Weight/Height")
plt.tight_layout()
plt.show()










```

```python
positions = data["Positions"].str.split(",").explode().str.strip()

pos_counts = positions.value_counts()
common = pos_counts[pos_counts >= 200]

# plot
plt.bar(common.index, common.values, color="orange")
plt.title("Frequently Used Player Positions (≥200 appearances)")
plt.xlabel("Player Positions")
plt.ylabel("Frequency")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
clubs_rate = data.groupby("Club")["Overall Rating"].mean().sort_values()

good_clubs = clubs_rate[clubs_rate > 75].head(30)
poor_clubs = clubs_rate[clubs_rate < 65].head(30)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))


# Left plot: Good clubs
axes[0].barh(good_clubs.index, good_clubs.values, color="green")
axes[0].set_title("Top 30 Clubs by Average Overall Rating above 75")
axes[0].set_xlabel("Average Overall Rating")
axes[0].set_ylabel("Club")

# Right plot: Poor clubs
axes[1].barh(poor_clubs.index, poor_clubs.values, color="red")
axes[1].set_title("Bottom 30 Clubs by Average Overall Rating below 60")
axes[1].set_xlabel("Average Overall Rating")
axes[1].set_ylabel("Club")

plt.tight_layout()
plt.show()





```

```python
Career_length = data["Retirment_year"] - data["Initial_year"]

plt.scatter(data["Initial_year"],Career_length,color="gold",marker="d",label="Career Span",alpha=0.7)
plt.xlabel("Initial Year")
plt.ylabel("Career Length (Years)")
plt.title("Initial Year vs Career Length")
plt.legend()
plt.grid(True)
plt.show()

```

```python
name_col = "Player Name"
total_col = "Total Stats"

top35 = data.nlargest(35, total_col)

# sort for better barh visualization
top35 = top35.sort_values(total_col)

plt.figure(figsize=(10, 12))
plt.barh(top35[name_col], top35[total_col])
plt.title("Top 35 Players by Total Stats")
plt.xlabel("Total Stats")
plt.ylabel("Player")
plt.tight_layout()
plt.show()
```
